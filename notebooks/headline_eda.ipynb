## 2. Descriptive Statistics

```python
import numpy as np

# Identify text columns (headlines, titles, content)
text_columns = df.select_dtypes(include=['str']).columns.tolist()
print("Text columns found:", text_columns)

# For each text column, calculate character count statistics
for col in text_columns:
    if col in df.columns and df[col].dtype == 'str':
        char_counts = df[col].astype(str).str.len()
        word_counts = df[col].astype(str).str.split().str.len()
        
        print(f"\n--- {col} Statistics ---")
        print(f"Character Count - Mean: {char_counts.mean():.2f}, Median: {char_counts.median():.2f}, Std: {char_counts.std():.2f}")
        print(f"Character Count - Min: {char_counts.min()}, Max: {char_counts.max()}")
        print(f"Word Count - Mean: {word_counts.mean():.2f}, Median: {word_counts.median():.2f}, Std: {word_counts.std():.2f}")
        print(f"Word Count - Min: {word_counts.min()}, Max: {word_counts.max()}")
```

```python
# Visualize text length distributions
import seaborn as sns

fig, axes = plt.subplots(len(text_columns), 2, figsize=(14, 4*len(text_columns)))
if len(text_columns) == 1:
    axes = [axes]

for idx, col in enumerate(text_columns):
    char_counts = df[col].astype(str).str.len()
    word_counts = df[col].astype(str).str.split().str.len()
    
    sns.histplot(char_counts, kde=True, ax=axes[idx][0])
    axes[idx][0].set_title(f'{col} - Character Count Distribution')
    axes[idx][0].set_xlabel('Character Count')
    
    sns.histplot(word_counts, kde=True, ax=axes[idx][1], color='orange')
    axes[idx][1].set_title(f'{col} - Word Count Distribution')
    axes[idx][1].set_xlabel('Word Count')

plt.tight_layout()
plt.show()
```

## 3. Publisher Analysis

```python
# Identify publisher columns (source, publisher, author, etc.)
publisher_columns = [col for col in df.columns if any(term in col.lower() for term in ['source', 'publisher', 'author', 'email', 'origin'])]
print("Potential publisher columns:", publisher_columns)

# If publisher columns exist, analyze them
if publisher_columns:
    for pub_col in publisher_columns:
        print(f"\n--- {pub_col} Analysis ---")
        pub_counts = df[pub_col].value_counts()
        print(f"Total unique values: {df[pub_col].nunique()}")
        print(f"\nTop 20 {pub_col}:")
        print(pub_counts.head(20))
        
        # Extract domains if email-like values exist
        email_like = df[pub_col].astype(str).str.contains('@', na=False).any()
        if email_like:
            print(f"\nEmail-like values detected in {pub_col}")
            domains = df[pub_col].astype(str).str.extract(r'@(.+)')[0].dropna().value_counts()
            print(f"Top domains:")
            print(domains.head(10))
else:
    print("No explicit publisher columns found. Checking all columns for publisher info...")
```

```python
# Visualize publisher distribution
if publisher_columns:
    for pub_col in publisher_columns:
        top_n = min(15, df[pub_col].nunique())
        pub_counts = df[pub_col].value_counts().head(top_n)
        
        plt.figure(figsize=(12, 6))
        pub_counts.plot(kind='barh')
        plt.title(f'Top {top_n} {pub_col}')
        plt.xlabel('Article Count')
        plt.ylabel(pub_col)
        plt.tight_layout()
        plt.show()
```

## 4. Publication Date Analysis

```python
# Identify date columns
date_columns = [col for col in df.columns if any(term in col.lower() for term in ['date', 'time', 'published', 'created'])]
print("Potential date columns:", date_columns)

# Parse and analyze date columns
date_col = None
for col in date_columns:
    try:
        df[f'{col}_parsed'] = pd.to_datetime(df[col])
        date_col = f'{col}_parsed'
        print(f"\nSuccessfully parsed {col}")
        print(f"Date range: {df[date_col].min()} to {df[date_col].max()}")
        print(f"Total days span: {(df[date_col].max() - df[date_col].min()).days}")
        break
    except:
        continue

if date_col:
    # Extract date/time features
    df['date_only'] = df[date_col].dt.date
    df['hour'] = df[date_col].dt.hour
    df['day_of_week'] = df[date_col].dt.day_name()
    df['month'] = df[date_col].dt.month
    df['year'] = df[date_col].dt.year
    
    print("\n--- Article Distribution by Time ---")
    print(f"Articles per day: {df.groupby('date_only').size().describe()}")
    print(f"\nArticles per hour (24h):\n{df['hour'].value_counts().sort_index()}")
    print(f"\nArticles per day of week:\n{df['day_of_week'].value_counts()}")
else:
    print("No valid date columns found")
```

```python
# Time series visualizations
if date_col:
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    
    # Articles over time
    daily_counts = df.groupby('date_only').size()
    axes[0, 0].plot(daily_counts.index, daily_counts.values, linewidth=1)
    axes[0, 0].set_title('Article Volume Over Time')
    axes[0, 0].set_xlabel('Date')
    axes[0, 0].set_ylabel('Number of Articles')
    axes[0, 0].grid(True, alpha=0.3)
    
    # Hour of day distribution
    hourly_counts = df['hour'].value_counts().sort_index()
    axes[0, 1].bar(hourly_counts.index, hourly_counts.values, color='steelblue')
    axes[0, 1].set_title('Articles by Hour of Day')
    axes[0, 1].set_xlabel('Hour (0-23)')
    axes[0, 1].set_ylabel('Count')
    axes[0, 1].grid(True, alpha=0.3, axis='y')
    
    # Day of week distribution
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    dow_counts = df['day_of_week'].value_counts().reindex(day_order)
    axes[1, 0].bar(range(len(dow_counts)), dow_counts.values, color='coral')
    axes[1, 0].set_xticks(range(len(dow_counts)))
    axes[1, 0].set_xticklabels([d[:3] for d in dow_counts.index], rotation=0)
    axes[1, 0].set_title('Articles by Day of Week')
    axes[1, 0].set_ylabel('Count')
    axes[1, 0].grid(True, alpha=0.3, axis='y')
    
    # Monthly distribution
    monthly_counts = df['month'].value_counts().sort_index()
    axes[1, 1].bar(monthly_counts.index, monthly_counts.values, color='mediumseagreen')
    axes[1, 1].set_title('Articles by Month')
    axes[1, 1].set_xlabel('Month')
    axes[1, 1].set_ylabel('Count')
    axes[1, 1].grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
```

## 5. Text Analysis - Keywords & Topics

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required NLTK data
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Prepare text data - combine all text columns
all_text = df[text_columns].fillna('').astype(str).agg(' '.join, axis=1)
print(f"Total text samples: {len(all_text)}")
print(f"\nFirst sample:\n{all_text.iloc[0][:200]}...")
```

```python
# TF-IDF Analysis to find important keywords
stop_words = set(stopwords.words('english'))

tfidf = TfidfVectorizer(max_features=100, stop_words='english', min_df=2, max_df=0.8)
tfidf_matrix = tfidf.fit_transform(all_text)

# Get top TF-IDF terms
feature_names = tfidf.get_feature_names_out()
tfidf_scores = tfidf_matrix.sum(axis=0).A1
top_indices = np.argsort(tfidf_scores)[-30:][::-1]

print("=== Top 30 Keywords by TF-IDF Score ===")
for idx in top_indices:
    print(f"{feature_names[idx]}: {tfidf_scores[idx]:.4f}")

# Visualize top keywords
fig, ax = plt.subplots(figsize=(12, 8))
top_20 = top_indices[:20]
ax.barh(range(len(top_20)), tfidf_scores[top_20])
ax.set_yticks(range(len(top_20)))
ax.set_yticklabels([feature_names[i] for i in top_20])
ax.set_xlabel('TF-IDF Score')
ax.set_title('Top 20 Keywords (TF-IDF)')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# CountVectorizer - Raw term frequencies
count_vec = CountVectorizer(max_features=100, stop_words='english', min_df=2, max_df=0.8)
count_matrix = count_vec.fit_transform(all_text)

# Get most frequent terms
feature_names_cv = count_vec.get_feature_names_out()
count_scores = count_matrix.sum(axis=0).A1
top_indices_cv = np.argsort(count_scores)[-30:][::-1]

print("\n=== Top 30 Most Frequent Terms ===")
for idx in top_indices_cv:
    print(f"{feature_names_cv[idx]}: {int(count_scores[idx])}")

# Visualize term frequencies
fig, ax = plt.subplots(figsize=(12, 8))
top_20_cv = top_indices_cv[:20]
ax.barh(range(len(top_20_cv)), count_scores[top_20_cv], color='steelblue')
ax.set_yticks(range(len(top_20_cv)))
ax.set_yticklabels([feature_names_cv[i] for i in top_20_cv])
ax.set_xlabel('Frequency')
ax.set_title('Top 20 Most Frequent Terms')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Latent Dirichlet Allocation (LDA) for Topic Modeling
n_topics = 5
print(f"\n=== LDA Topic Modeling (n_topics={n_topics}) ===\n")

lda = LatentDirichletAllocation(n_components=n_topics, random_state=42, max_iter=10)
lda.fit(count_matrix)

# Display topics
for topic_idx, topic in enumerate(lda.components_):
    top_features_idx = topic.argsort()[-10:][::-1]
    top_features = [feature_names_cv[i] for i in top_features_idx]
    weights = topic[top_features_idx]
    
    print(f"Topic {topic_idx + 1}:")
    for feature, weight in zip(top_features, weights):
        print(f"  {feature}: {weight:.4f}")
    print()

# Get topic distribution for first 10 documents
doc_topics = lda.transform(count_matrix)
print("=== Document-Topic Distribution (first 5 documents) ===")
print(f"Shape: {doc_topics.shape}")
print(f"\nFirst 5 documents:\n{pd.DataFrame(doc_topics[:5], columns=[f'Topic {i+1}' for i in range(n_topics)]).round(4)}")
```

```python
# Visualize topic distribution
dominant_topics = doc_topics.argmax(axis=1)
topic_dist = pd.Series(dominant_topics).value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Topic distribution
axes[0].bar(topic_dist.index + 1, topic_dist.values, color='teal')
axes[0].set_xlabel('Topic')
axes[0].set_ylabel('Number of Documents')
axes[0].set_title('Distribution of Dominant Topics')
axes[0].set_xticks(range(1, n_topics + 1))

# Topic heatmap (average weight per topic across all documents)
avg_topic_dist = doc_topics.mean(axis=0)
axes[1].bar(range(1, n_topics + 1), avg_topic_dist, color='coral')
axes[1].set_xlabel('Topic')
axes[1].set_ylabel('Average Weight')
axes[1].set_title('Average Topic Weights Across All Documents')
axes[1].set_xticks(range(1, n_topics + 1))

plt.tight_layout()
plt.show()

# Add dominant topic to dataframe
df['dominant_topic'] = dominant_topics + 1
print("\n=== Articles per Topic ===")
print(df['dominant_topic'].value_counts().sort_index())
```

```python
# Detect recurring themes and specific keywords
recurring_themes = {
    'FDA/Regulatory': [r'\bfda\b', r'\breulatory', r'\bapproval\b', r'\bclinical trial', r'\bfda approval'],
    'Price/Valuation': [r'\bprice target', r'\bstock price', r'\bvaluation', r'\bupgrade', r'\bdowngrade', r'\boverweight', r'\bunderweight'],
    'Earnings': [r'\bearning', r'\beps\b', r'\bevenue\b', r'\bquarterly\b', r'\bq[1-4]\b', r'\beat', r'\bmiss'],
    'Merger/Acquisition': [r'\bmerger\b', r'\bacquisition\b', r'\bacquire', r'\btakeover', r'\bdeal\b'],
    'Technology': [r'\bai\b', r'\bartificial intelligence', r'\bdata', r'\bcloud', r'\bsoftware', r'\bdigital'],
    'Market/Economic': [r'\bmarket\b', r'\beconomy\b', r'\brecession\b', r'\binflation\b', r'\binterest rate'],
}

print("=== Recurring Themes Detection ===\n")
for theme, patterns in recurring_themes.items():
    combined_pattern = '|'.join(f'({p})' for p in patterns)
    matching_articles = all_text.str.lower().str.contains(combined_pattern, regex=True, na=False).sum()
    pct = (matching_articles / len(all_text)) * 100
    print(f"{theme}: {matching_articles} articles ({pct:.2f}%)")

# Count individual significant keywords/phrases
print("\n=== Specific Keywords/Phrases Frequency ===\n")
keywords_to_track = [
    ('FDA Approval', r'\bfda approval'),
    ('Price Target', r'\bprice target'),
    ('Earnings Beat', r'\bearnings? beat'),
    ('Analyst Rating', r'\banalyst|rating'),
    ('AI/Machine Learning', r'\b(?:ai|artificial intelligence|machine learning)\b'),
    ('Cloud Computing', r'\bcloud'),
    ('Cybersecurity', r'\bcybersecurity|security breach'),
]

for keyword, pattern in keywords_to_track:
    count = all_text.str.lower().str.contains(pattern, regex=True, na=False).sum()
    pct = (count / len(all_text)) * 100
    print(f"{keyword}: {count} ({pct:.2f}%)")
```
